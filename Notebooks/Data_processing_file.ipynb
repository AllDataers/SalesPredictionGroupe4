***Importation des modules***

```python
import pandas as pd
import numpy as np
```

```python
df_2019 = pd.read_csv("https://raw.githubusercontent.com/CerateAGNIKPE/SalesPredictionGroupe4/C%C3%A9rate/dataset/Sales_April_2019.csv")
print("Data Information:\n")
print(df_2019.info())


df_2019.head()

```

```python
# Check for missing values
print("\nMissing values:\n")
print(df_2019.isnull().sum())
```

```python
df_2019 = df_2019.dropna()
```

```python
df_2019.head()
```

```python
print("\nMissing values:\n")
print(df_2019.isnull().sum())
```

```python
df_2019.nunique().to_frame(name='Count of unique values')
```

```python
df_2019['Product'].value_counts().to_frame(name='Quantity for each product')
```

```python
df_2019.groupby("Product")['Price Each'].first().to_frame('Price of product')
```

*On note la présence d'un produit appelé produit et sont prix est price earch , donc nous allons sélectionner par la suite les lignes dont le produit est Product*

```python
df_2019[df_2019['Product']=='Product']
```

*Comme le montre il y a encore des erreurs dans la base à supprimer, nous allons supprimer les lignes dont la colonne Product est égale à Product*

```python
df_2019.loc[519]
```

```python
# Sélectionner les indices des lignes où la colonne 'Product' a la valeur 'Product'
indices_to_drop = df_2019[df_2019['Product'] == 'Product'].index

# Supprimer les lignes correspondantes
df_2019.drop(indices_to_drop, inplace=True)

# Afficher le DataFrame après la suppression
df_2019.head()
```

```python
df_2019['Product'].value_counts().to_frame(name='Quantity for each product')
```

```python
df_2019.groupby("Product")['Price Each'].first().to_frame('Price each product')
```
