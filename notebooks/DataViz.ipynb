```python
df2 = pd.read_excel('df_feb2019.xlsx')
df3 = pd.read_excel('df_mar2019.xlsx')
df4= pd.read_excel('df_apri2019.xlsx')
df5 = pd.read_excel('df_may2019.xlsx')
df6 = pd.read_excel('df_jun2019.xlsx')
df7 = pd.read_excel('df_jul2019.xlsx')
df8 = pd.read_excel('df_aug2019.xlsx')
df9 = pd.read_excel('df_sep2019.xlsx')
df10 = pd.read_excel('df_oct2019.xlsx')
df11 = pd.read_excel('df_nov2019.xlsx')
df12 = pd.read_excel('df_dec2019.xlsx')
```

```python
# Fusionner les DataFrames en un seul
df = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12], ignore_index=True)

# Affichage du DataFrame rsultant
df.head()
```

```python
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.io import output_notebook
```

```python
# Groupement par produit et somme de la quantit, tri par ordre dcroissant
df_grouped = df.groupby('Product')['QuantityOrdered'].sum().reset_index()
df_grouped = df_grouped.sort_values(by='QuantityOrdered', ascending=True)

# Crer une source de donnes Bokeh
source = ColumnDataSource(df_grouped)

# Crer un diagramme en barres avec Bokeh (inverser l'axe)
p = figure(y_range=df_grouped['Product'], height=400, width=600, title="Total Quantity Ordered per Product",
           toolbar_location="right", tools="pan,box_zoom,wheel_zoom,reset,save", x_axis_label="Total Quantity Ordered", y_axis_label="Product")
p.grid.visible = False

# Crer le diagramme en barres
p.hbar(y='Product', right='QuantityOrdered', source=source, height=0.4, color="blue")


# Afficher le diagramme en barres avec Bokeh dans le notebook
output_notebook()
show(p)
```
