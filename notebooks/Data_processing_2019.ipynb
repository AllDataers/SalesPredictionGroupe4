**GENERATION D'UNE FONCTION**

```python
import pandas as pd

def load_and_clean_data(url):
    # Charger les donn√©es depuis l'URL
    df = pd.read_csv(url)

    # Renommer les colonnes
    df.rename(columns={
        'Order ID': 'OrderID',
        'Quantity Ordered': 'QuantityOrdered',
        'Price Each': 'PriceEach',
        'Order Date': 'OrderDate',
        'Purchase Address': 'PurchaseAddress'
    }, inplace=True)
    
    # Supprimer les lignes avec des valeurs manquantes
    df = df.dropna()

    # Trouver les indices des lignes o√π la colonne 'Product' a la valeur 'Product'
    indices_to_drop = df[df['Product'] == 'Product'].index

    # Supprimer les lignes correspondantes
    df.drop(indices_to_drop, inplace=True)
    
    df['QuantityOrdered'] = df['QuantityOrdered'].astype(int)
    df['PriceEach'] = df['PriceEach'].astype(float)
    df['OrderDate'] = pd.to_datetime(df['OrderDate'])

    df['Sales']= df['QuantityOrdered'] * df['PriceEach'] # Chiffre d'affaire
    df['City'] = df['PurchaseAddress'].str.split(',').str[1].str.strip()
    df['Month'] = df['OrderDate'].dt.month_name()
    df['Hour'] = df['OrderDate'].dt.hour

    return df

# Utilisation de la fonction avec l'URL fournie
url = "https://raw.githubusercontent.com/AllDataers/SalesPredictionGroupe4/main/data/raw/Sales_August_2019.csv"
df_aug = load_and_clean_data(url)


```

```python
df_aug.head(1000)
```

```python
df_aug.to_csv('df_aug2019.csv', index=False)

print("DataFrame export√© avec succ√®s au format csv.")
```
